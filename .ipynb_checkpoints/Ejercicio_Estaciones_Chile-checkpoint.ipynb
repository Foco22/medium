```python
#trato de hacer una rutina eh
def anexo(merge,nuevo):
    datos_nnn = pd.read_csv("./LACA_blended_custom/TG_STAID000"+nuevo+".csv")
    datos_nnn.columns = [column_name.replace(' ','') for column_name in datos_nnn.columns]
    datos_nnn['DATE'] = pd.to_datetime(datos_nnn['DATE'],format='%Y%m%d')
    datos_nnn.set_index('DATE',inplace=True)
    merge_n=pd.merge(merge,datos_nnn, how='inner', left_index=True, right_index=True)
    borrar = ['SOUID','Q_TG']
    merge_n.drop(borrar, axis=1, inplace=True)
    merge_n.rename(columns={'TG':'TG'+nuevo}, inplace=True)
    return merge_n
```

```python
print ('Inicio de Prediccciones')
def prediccion(X_test, y_test):
    y_predict = ridge_model_full_data.predict(X_test)
    rms_ridge = round(rmsle(y_test, y_predict ),4)
    acc_ridge = round(ridge_model_full_data.score(X_test, y_test)*100, 2)

    y_predict = lasso_model_full_data.predict(X_test)
    rms_lasso = round(rmsle(y_test, y_predict ),4)
    acc_lasso = round(lasso_model_full_data.score(X_test, y_test)*100, 2)

    y_predict = elastic_model_full_data.predict(X_test)
    rms_elastic = round(rmsle(y_test, y_predict ),4)
    acc_elastic = round(elastic_model_full_data.score(X_test, y_test)*100, 2)

    y_predict = svr_model_full_data.predict(X_test)
    rms_svr = round(rmsle(y_test, y_predict ),4)
    acc_svr = round(svr_model_full_data.score(X_test, y_test)*100, 2)

    y_predict = xgb_model_full_data.predict(X_test)
    rms_xgb = round(rmsle(y_test, y_predict ),4)
    acc_xgb = round(xgb_model_full_data.score(X_test, y_test)*100, 2)

    y_predict = gbr_model_full_data.predict(X_test)
    rms_gbr = round(rmsle(y_test, y_predict ),4)
    acc_gbr = round(gbr_model_full_data.score(X_test, y_test)*100, 2)

    y_predict = lgb_model_full_data.predict(X_test)
    rms_lgb = round(rmsle(y_test, y_predict ),4)
    acc_lgb = round(lgb_model_full_data.score(X_test, y_test)*100, 2)

    y_predict = lr_model_full_data.predict(X_test)
    rms_lr = round(rmsle(y_test, y_predict ),4)
    acc_lr = round(lr_model_full_data.score(X_test, y_test)*100, 2)

    y_predict = rf_model_full_data.predict(X_test)
    rms_rf = round(rmsle(y_test, y_predict ),4)
    acc_rf = round(rf_model_full_data.score(X_test, y_test)*100, 2)

    models = pd.DataFrame({
    'Model': ['Ridge','Lasso','Elastic','SVR', 'XGB','GBR','Ligth GB', 
              'Linear','Random Forest'],
    'Score': [acc_ridge, acc_lasso, acc_elastic, acc_svr, 
              acc_xgb, acc_gbr, acc_lgb, acc_lr, acc_rf],
    'RMSLE': [rms_ridge, rms_lasso, rms_elastic, rms_svr, 
              rms_xgb, rms_gbr, rms_lgb, rms_lr, rms_rf],})
    models = models.sort_values(by='Score', ascending=False)
    return models
    
```
